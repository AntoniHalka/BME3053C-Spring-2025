# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python

```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return bmi

def main():
    #input weight in kilograms and height in meters
    try:
        weight = float(input("Enter your weight in kilograms: "))
        height = float(input("Enter your height in meters: "))
        
        # Calculate BMI
        bmi = calculate_bmi(weight, height)
        
        # Print the result
        print(f"Your BMI is: {bmi:.2f}")

        # Determine the BMI category
        if bmi < 18.5:
            category = "Underweight"
        elif 18.5 <= bmi < 24.9:
            category = "Normal weight"
        elif 25 <= bmi < 29.9:
            category = "Overweight"
        else:
            category = "Obesity"
        print(f"You are classified as: {category}")
    except ValueError as e:
        # Handle invalid input
        print(f"Invalid input: {e}")        

    main(30, 1.75)  # Example usage with weight=30kg and height=1.75m
````

   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
### ✏️ Exercise Create a function that processes medical image data:
   #- Accept a numpy array representing an image
   #- Apply basic image processing (normalization, filtering)
   #- Return the processed image
   #- Include proper error handling
import numpy as np
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt  # Add this import
def process_medical_image(image):
    """Process a medical image represented as a numpy array."""
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    
    if image.ndim != 2:
        raise ValueError("Image must be a 2D array.")
    
    # Normalize the image
    image_min = np.min(image)
    image_max = np.max(image)
    normalized_image = (image - image_min) / (image_max - image_min)
    
    # Apply Gaussian filter for smoothing
    processed_image = gaussian_filter(normalized_image, sigma=1)
    
    return processed_image
# Test the function with a sample image
if __name__ == "__main__":
    # Create a sample image (2D numpy array)
    sample_image = np.random.rand(100, 100) * 255
    processed_image = process_medical_image(sample_image)
    
    # Display the original and processed images
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(sample_image, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title("Processed Image")
    plt.imshow(processed_image, cmap='gray')
    
    plt.show()
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecord:
    def __init__(self):
        self.records = []

    def add_record(self, name, age, weight, height):
        """Add a new patient record."""
        if not isinstance(name, str) or not name:
            raise ValueError("Name must be a non-empty string.")
        if age <= 0:
            raise ValueError("Age must be positive.")
        if weight <= 0:
            raise ValueError("Weight must be positive.")
        if height <= 0:
            raise ValueError("Height must be positive.")
        
        record = {
            "name": name,
            "age": age,
            "weight": weight,
            "height": height
        }
        self.records.append(record)
        return f"Record for {name} added successfully."

    def update_record(self, name, age=None, weight=None, height=None):
        """Update an existing patient record."""
        for record in self.records:
            if record["name"] == name:
                if age is not None:
                    if age <= 0:
                        raise ValueError("Age must be positive.")
                    record["age"] = age
                if weight is not None:
                    if weight <= 0:
                        raise ValueError("Weight must be positive.")
                    record["weight"] = weight
                if height is not None:
                    if height <= 0:
                        raise ValueError("Height must be positive.")
                    record["height"] = height
                return f"Record for {name} updated successfully."
        return f"No record found for {name}."

    def get_record(self, name):
        """Retrieve a patient record."""
        for record in self.records:
            if record["name"] == name:
                return record
        return f"No record found for {name}."

    def calculate_average_weight(self):
        """Calculate the average weight of all patients."""
        if not self.records:
            return "No records available."
        total_weight = sum(record["weight"] for record in self.records)
        average_weight = total_weight / len(self.records)
        return round(average_weight, 2)

    def calculate_average_height(self):
        """Calculate the average height of all patients."""
        if not self.records:
            return "No records available."
        total_height = sum(record["height"] for record in self.records)
        average_height = total_height / len(self.records)
        return round(average_height, 2)

    def calculate_bmi(self):
        """Calculate BMI for all patients."""
        if not self.records:
            return "No records available."
        bmi_list = []
        for record in self.records:
            bmi = record["weight"] / (record["height"] ** 2)
            bmi_list.append(round(bmi, 2))
        return bmi_list

# Example usage
if __name__ == "__main__":
    patient_record = PatientRecord()
    print(patient_record.add_record("Alice", 30, 65, 1.7))
    print(patient_record.add_record("Bob", 25, 75, 1.8))
    print(patient_record.update_record("Alice", weight=68))
    print(patient_record.get_record("Alice"))
    print(f"Average Weight: {patient_record.calculate_average_weight()} kg")
    print(f"Average Height: {patient_record.calculate_average_height()} m")
    print(f"BMIs: {patient_record.calculate_bmi()}")
````
